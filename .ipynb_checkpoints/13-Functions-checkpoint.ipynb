```python
##args , kwargs(arguments, key word arguments)
```

```python
def argsSum(*args):
    return sum(args)
```

```python
argsSum(10,20,30,40)
```

```python
def args_example(*args):
    print(args)
```

```python
args_example(10,1413,213312,239128371)
```

```python
def kwargs_example(**kwargs):
    print(kwargs)
```

```python
kwargs_example(apple=100,banana=150,melon=200)   ##sozluk yapmis oldu
```

```python
def kwargs_example_2(**kwargs):
    if "apple" in kwargs:
        print("appleee")
    else:
        print(":(")
```

```python
kwargs_example_2(apple=100,banana=150)
```

```python
kwargs_example_2(banana=150,melon=200)
```
