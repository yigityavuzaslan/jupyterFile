BREAK, CONTUNUE, PASS 

```python
myList=[10,20,30,40,50,60,70]
```

```python
for number in myList:                               ##arkada 50 60 70 i deniyo 
    print(number)                                   ##40 olunca nasil durduracagiz
    if number==40:
        print("yes")
```

```python
for number in myList:                               
    print(number)                                  
    if number==40:
        print("yes")
        break
```

```python
myList=[10,20,30,40,50,60,70]
```

```python
for num in myList:                           ##dur ve donguden cik devam et demektir
    print(num)
    if num==40:
        continue
        print("yes")
        
```

```python
for num in myList:                          
    if num==40:
        continue
    print(num)
```

```python
for num in myList:                          
    pass                                     ##sonrasinda duzeltmek icin kullanilir genelde
```
