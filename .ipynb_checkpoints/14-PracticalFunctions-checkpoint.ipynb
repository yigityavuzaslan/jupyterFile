```python
def divideNumber(number):
    return number/2
```

```python
divideNumber(20)
```

```python
myList=[3,5,7,10,20,30]
```

```python
myResultList=[]
```

```python
for num in myList:
    myResultList.append(divideNumber(num))
print(myResultList)
```

```python
myResultList
```

```python
def controlString(string):
    return "Yigit" in string
```

```python
controlString("Yigit Yavuzaslan")
```

```python
myStringList=["Yigit","Yigit Yavuzaslan","Yavuzaslan","Aslan"]
```

```python
list(map(controlString,myStringList))
```

##LEGB -> L: LOCAL, E:Enclosing, G:GLOBAL, B:Built-In

```python
#global
myString="yigit"
def myFunction():
    #enclosing
    myString="yigit 2"
    print(myString)
    def myFunction2():
        #local
        myString="yigit 3"
        print(myString)
    myFunction2()
```

```python
myString
```

```python
myFunction()
```

```python
def test1():
    myVariable=10
    print(myVariable*2)
    
def test2():
    print(myVariable*3)
```

```python
y=10
def newFunction(y):
    print(y)
    y=5
    print(y)
    return y
```

```python
newFunction(15)
```

```python
y=10
def changeY():
    global y
    y=5
    print(y)      ##global y yi kullanacam dedik
```
